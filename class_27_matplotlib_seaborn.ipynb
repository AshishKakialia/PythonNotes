```python
import matplotlib.pyplot as plt
```

```python
# pie chart

exp_val = [1400, 800, 300, 410, 250]
exp_label = ['house rent', 'food', 'phone', 'car', 'other utilities']
plt.pie(exp_val, labels = exp_label)
```

```python
exp_val = [1400, 800, 300, 410, 250]
exp_label = ['house rent', 'food', 'phone', 'car', 'other utilities']
plt.pie(exp_val, labels=exp_label, autopct='%0.2f%%')
```

```python
exp_val = [1400, 800, 300, 410, 250]
exp_label = ['house rent', 'food', 'phone', 'car', 'other utilities']
plt.pie(exp_val, labels=exp_label, autopct='%0.2f%%',  radius=1.4)
```

```python
# explode

exp_val = [1400, 800, 300, 410, 250]
exp_label = ['house rent', 'food', 'phone', 'car', 'other utilities']
plt.pie(exp_val, labels=exp_label, autopct='%0.0f%%',  radius=1.4 , explode = [0.02,0,0,0,0.08])
```

```python
# scatter plot

x = [1,2,3,4,5,6,7,8]
y = [5,2,5,7,8,4,2,5]

plt.scatter(x,y)
```

```python
x = [1,2,3,4,5,6,7,8]
y = [i*2 for i in x]

plt.scatter(x,y)
```

```python
# sav fig

exp_val = [1400, 800, 300, 410, 250]
exp_label = ['house rent', 'food', 'phone', 'car', 'other utilities']
plt.pie(exp_val, labels=exp_label, autopct='%0.0f%%',  radius=1.4 , explode = [0.02,0,0,0,0.08])
plt.savefig('piechart.png')
```

```python
exp_val = [1400, 800, 300, 410, 250]
exp_label = ['house rent', 'food', 'phone', 'car', 'other utilities']
plt.pie(exp_val, labels=exp_label, autopct='%0.0f%%',  radius=1.4 , explode = [0.02,0,0,0,0.08])
plt.savefig('piechart.pdf')
```

# seaborn

```python
import seaborn as sns
```

```python
tips = sns.load_dataset('tips')
tips
```

```python
sns.relplot(x='total_bill', y='tip', data=tips)
```

```python
sns.relplot(x='total_bill', y='tip', data=tips , hue = 'smoker') # Hue- categorize based on color
```

```python
sns.relplot(x='total_bill', y='tip', data=tips , hue='smoker', style='time' ) # baes on the different styles
```

```python
sns.relplot(x='total_bill', y='tip', data=tips , hue='smoker', style='time', size='sex' )
```

```python
sns.relplot(x='total_bill', y='tip', data=tips , hue='smoker', col='time', size='sex' )
```

```python
sns.catplot(x='day', y='total_bill', data=tips)
```

```python
sns.catplot(x='day', y='total_bill', data=tips, jitter=False)
```

```python
sns.catplot(x='day', y='total_bill', data=tips, kind='swarm')
```

```python
sns.catplot(x='day', y='total_bill', data=tips, kind='violin')
```

```python
# box plot

sns.catplot(x='day', y='total_bill', data=tips, kind='box')
```

```python
# distribution plot

sns.distplot(tips['total_bill'])
```

```python
sns.jointplot(x=tips['total_bill'], y=tips['tip'])
```

```python
sns.jointplot(x=tips['total_bill'], y=tips['tip'], kind='hex')
```

```python
sns.jointplot(x=tips['total_bill'], y=tips['tip'], kind='kde')
```

```python
# pairplot

sns.pairplot(tips)
```
